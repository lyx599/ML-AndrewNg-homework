```python
import scipy.io as sio
import numpy as np
from sklearn.metrics import classification_report#这个包是评价报告
```

```python
sio.loadmat('ex3weights.mat')
```

```python
def load_weight(path):
    data = sio.loadmat(path)
    return data['Theta1'],data['Theta2']
```

```python
theta1,theta2 = load_weight('ex3weights.mat')
theta1.shape,theta2.shape
```

```python
#加载数据
def load_data(path,transpose=True):
    data = sio.loadmat(path)
    y = data.get('y')
    y = y.reshape(y.shape[0])
    
    x = data.get('X')
    #转置把图像放正，原数据图像是错误的方向
    if transpose:
        #对每个图片放正
        x = np.array([im.reshape((20,20)).T for im in x])
        #再展开
        x = np.array([im.reshape(400) for im in x])
    
    return x,y
```

```python
#因此在数据加载函数中，原始数据做了转置，然而，转置的数据与给定的参数不兼容，
#因为这些参数是由原始数据训练的。 所以为了应用给定的参数，我需要使用原始数据（不转置）
x,y = load_data('ex3data1.mat',transpose=False)
m = x.shape[0]
x = np.column_stack((np.ones(m),x)) 
x.shape,y.shape
```

# 神经网络模型图示

# feed forward prediction（前馈预测）

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
# theta1 (25, 401)
# theta2 (10, 26)
a1 = x 
```

```python
z2 = a1 @ theta1.T  # a1(5000,401) @ theta.T(401,25) =(5000,25)
z2.shape
```

```python
a2 = sigmoid(z2)
a2 = np.column_stack((np.ones(m),a2))
z3 = a2 @ theta2.T # a2(5000,26) @ theta2.T(26,10) = (5000,10)
z3.shape
```

```python
a3 = sigmoid(z3)
a3
```

```python
# 因为索引从0开始 对每个索引加1就得到正确的值
y_pred = np.argmax(a3, axis=1) + 1  
y_pred.shape
```

```python
print(classification_report(y, y_pred))
```
