# 1-异常检测

在本练习中，我们将使用高斯模型实现异常检测算法，并将其应用于检测网络上的故障服务器。

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import scipy.io as sio
```

## 1.1高斯分布

我们的第一个任务是使用高斯模型来检测数据集中未标记的示例是否应被视为异常。 我们有一个简单的二维数据集开始，以帮助可视化该算法正在做什么。 

```python
raw_data = sio.loadmat('data/ex8data1.mat')
raw_data
```

```python
raw_data.keys()
```

```python
x = raw_data['X']
x.shape
```

```python
fig, ax = plt.subplots(figsize=(12,8))
ax.scatter(x[:,0], x[:,1])
plt.show()
```

这是一个非常紧密的聚类，几个值远离了聚类。 在这个简单的例子中，这些可以被认为是异常的。 为了弄清楚，我们为数据中的每个特征估计高斯分布。 为此，我们将创建一个返回每个要素的均值和方差的函数。

## 1.2估计参数

```python
def estimate_gaussian(x):
    mu = x.mean(axis=0)
    sigam = x.std(axis=0)
    
    return mu,sigam
```

```python
mu,sigma = estimate_gaussian(x)
mu,sigma
```

## 1.3确定阈值 

现在我们有了我们的模型参数，我们需要确定概率阈值，来表明一个样本是否被认为是一个异常。 为此，我们需要使用一组标记的验证数据（其中真实异常样本已被标记），并在给出不同阈值的情况下，对模型的性能进行鉴定。

```python
#验证集
xval = raw_data['Xval']
yval = raw_data['yval']
xval.shape,yval.shape
```

我们还需要一种计算数据点正态分布的概率的方法。 使用SciPy内置的方法。

```python
from scipy.stats import norm
#loc是均值，scale是标准差，pdf是概率密度函数 具体参考官方文档
norm.pdf(mu[0],loc=mu[0],scale=sigma[0])
```

我们还可以将数组传递给概率密度函数，并获得数据集中那列特征每个点的概率密度。

```python
#训练数据集第一个特征所有点的概率 
norm.pdf(x[:,0],loc=mu[0],scale=sigma[0])                                       
```

```python
px = np.zeros((x.shape[0], x.shape[1]))
px[:,0] = norm.pdf(x[:,0],loc=mu[0],scale=sigma[0])
px[:,1] = norm.pdf(x[:,1],loc=mu[1],scale=sigma[1])
p_px = px.prod(axis=1)
p_px.shape
```

我们计算并保存给定上述的高斯模型参数的``验证集``中每个值的概率密度。

```python
pval = np.zeros((xval.shape[0],xval.shape[1]))
pval[:,0] = norm.pdf(xval[:,0],loc=mu[0],scale=sigma[0])
pval[:,1] = norm.pdf(xval[:,1],loc=mu[1],scale=sigma[1]) 
pval.shape
```

接下来，我们需要一个函数，找到给定概率密度值和真实标签的最佳阈值。 为了做到这一点，我们将为不同的epsilon值计算F1分数。

<img style="float: left;" src="images/F1.png">

```python
#验证集每个样例的概率
p_pval = pval.prod(axis=1)
p_pval.shape,p_pval.max(),p_pval.min()
```

```python
def select_threshold(pval,yval):
    best_epsilon = 0
    best_f1 = 0
    f1 = 0
    
    step = (pval.max() - pval.min()) / 1000
    
    for epsilon in np.arange(pval.min(),pval.max(),step):
        preds = pval < epsilon
        #因为逻辑与要shape相同，不然会出现问题
        preds = np.matrix(preds).T
        
        tp = np.sum(np.logical_and(preds == 1,yval == 1)).astype(float)
        fp = np.sum(np.logical_and(preds == 1, yval == 0)).astype(float)
        fn = np.sum(np.logical_and(preds == 0, yval == 1)).astype(float)
        
        precision = tp / (tp + fp)
        recall = tp / (tp + fn)
        f1 = (2 * precision * recall) / (precision + recall)
        
        if f1 > best_f1:
            best_f1 = f1
            best_epsilon = epsilon
    
    return best_epsilon, best_f1
```

```python
epsilon, f1 = select_threshold(p_pval, yval)
epsilon, f1
```

```python
outliers = np.where(p_px < epsilon)
outliers
```

```python
fig, ax = plt.subplots(figsize=(12,8))
ax.scatter(x[:,0], x[:,1])
ax.scatter(x[outliers[0],0], x[outliers[0],1], s=50, color='r', marker='o')
plt.show()
```

图中可以看到检测的异常点还是很合理的

## 1.4高维数据
>还是同样的几步操作

```python
raw_data = sio.loadmat('data/ex8data2.mat')
raw_data.keys()
```

```python
x = raw_data['X']
xval = raw_data['Xval']
yval = raw_data['yval']
x.shape,xval.shape,yval.shape
```

```python
mu,sigma = estimate_gaussian(x)
mu,sigma
```

```python
phx = np.zeros((x.shape[0], x.shape[1]))
for i in range(x.shape[1]):
    phx[:,i] = norm.pdf(x[:,i],loc=mu[i],scale=sigma[i])

p_phx = phx.prod(axis=1)
p_phx.shape
```

```python
phval = np.zeros((xval.shape[0],xval.shape[1]))
for i in range(xval.shape[1]):
    phval[:,i] = norm.pdf(xval[:,i],loc=mu[i],scale=sigma[i])
p_phval = phval.prod(axis=1)
p_phval.shape
```

```python
epsilon, f1 = select_threshold(p_phval, yval)
epsilon, f1
```

```python
outliers = np.where(p_phx < epsilon)
outliers[0].size
```

与期望输出的阈值1.38e-18和117异常点一致
