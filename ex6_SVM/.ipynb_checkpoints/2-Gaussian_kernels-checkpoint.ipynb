# 高斯核函数

```python
import matplotlib.pyplot as plt
from sklearn import svm
import numpy as np
import pandas as pd
import seaborn as sns
import scipy.io as sio
```

<img style="float: left;" src="images/gaussian.png">

```python
def gaussian_kernel(x1,x2,sigma):
    return np.exp(- np.power(x1 - x2,2).sum() / (2 * (sigma ** 2)))
```

## 示例数据2

```python
#观察数据 
raw_data = sio.loadmat('data/ex6data2.mat')
raw_data
```

## 可视化数据 

```python
data = pd.DataFrame(raw_data['X'],columns=['x1','x2'])
data['y'] = raw_data['y']
data.head()
```

```python
positive = data[data['y'].isin([1])]
negative = data[data['y'].isin([0])]

fig,ax = plt.subplots(figsize=(12,8))
ax.scatter(positive['x1'],positive['x2'],marker='x',label='positive')
ax.scatter(negative['x1'],negative['x2'],marker='o',label='negative')
plt.legend()
plt.show()
```

```python
svc = svm.SVC(C=100,kernel='rbf',gamma=10,probability=True)
svc
```

```python
svc.fit(data[['x1', 'x2']], data['y'])
svc.score(data[['x1', 'x2']], data['y'])
```

```python
predict_prob = svc.predict_proba(data[['x1','x2']])[:,0]
fig, ax = plt.subplots(figsize=(12,8))
ax.scatter(data['x1'], data['x2'], s=30, c=predict_prob, cmap='seismic')
plt.show()
```

可以看到和上面真实数据对比基本吻合，只有个别边界上的数据存在错误划分
